```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x1 = -2 * np.random.rand(50,2)
x2 = 1 + 2 * np.random.rand(50,2)
plt.scatter(x1,x2,s=50,c='b')
plt.show()
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = 'b')
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
centers = Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(-0.94665068, -0.97138368, s=200, c='g', marker='s') #with 2 centers
plt.scatter(2.01559419, 2.02597093, s=200, c='r', marker='s') #with 2 centers
plt.show()
```
